## Structuring input data using sliding windows
#### can change window_size to control # prior months considered to make prediction

https://www.geeksforgeeks.org/dsa/window-sliding-technique/

```python
# read in list of arrays for each pixel - these are fully filtered
sfe_pre = np.load("/scratch/users/ashdef/pre_treatment_data/no_nans/sfe_pre_arrays.npz", allow_pickle = True)["sfe_pre"] 

climate_pre = np.load("/scratch/users/ashdef/pre_treatment_data/no_nans/climate_pre_arrays.npz", allow_pickle = True)["clim_pre"]
```

```python
window_size = 3 # number of months prior to consider input data prior to current timestep - also tried 6 month window

# will store climate data for 3 months behind timestep, sfe data beginning at month 4
climate_window = []
sfe_window = []

# loop over pixels
for pixel in range(len(sfe_pre)):
    for t in range(window_size, len(sfe_pre[pixel])): # exclude timesteps before window size, et won't be predicted for these
        climate_window.append(climate_pre[pixel][t - window_size:t][:]) # array of 6 months before current timestep, for each clim var
        sfe_window.append(sfe_pre[pixel][t])
                      
climate_window = np.array(climate_window)
sfe_window = np.array(sfe_window)

# wrap the lists as object arrays - allows arrays of different lengths to be saved
sfe_obj = np.array(sfe_window, dtype=object)
climate_obj = np.array(climate_window, dtype=object)

np.savez_compressed("/scratch/users/ashdef/pre_treatment_data/no_nans/sfe_filtered_3mwindow.npz", sfe_window = sfe_obj)
np.savez_compressed("/scratch/users/ashdef/pre_treatment_data/no_nans/climate_filtered_3mwindow.npz", climate_window =climate_obj)
```
